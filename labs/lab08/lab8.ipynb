$$G = \sum_{k=1}^{K}p_{mk}(1-p_{mk}) $$

#### Question 7

You should find that the trees vary at least somewhat with each random split of your data. This is what we mean when we say decision trees are "unstable"; small variations in the data can lead to very different trees. Decision trees can therefore vary considerably in their quality depending on your random sample, and it can also be hard to learn about the importance of variables (defined as their contribution to reducing the Gini index in a given tree) from any single tree. 

However, as you created multiple trees you still likely found that some variables showed up more often than others. Perhaps this process of resampling our data and creating multiple trees could help us resolve this inherent instability. In fact, this is called **Bagging**.

Bootstrap your decision trees 1000 times, resampling your training data with replacement each time and fitting a tree. Calculate the average importance of each feature over the 1000 bootstrapped trees.

*Hint:* Decision trees have an attribute called `feature_importances_` 

```python
n_bootstraps = 1000
feature_importances = []

for i in range(n_bootstraps):
    indices = np.random.choice(X_train.index, size=len(X_train), replace=True)
    X_bootstrap = X_train.loc[indices]
    y_bootstrap = y_train.loc[indices]

    clf_bootstrap = DecisionTreeClassifier()
    clf_bootstrap.fit(X_bootstrap, y_bootstrap)

    feature_importances.append(clf_bootstrap.feature_importances_)
```

```python
mean_importances = np.mean(feature_importances, axis=0)

importance = pd.DataFrame({"feature": features.columns, "mean importance": mean_importances})
importance.sort_values("mean importance", ascending = False)
```

You should observe that ~5 features are the only ones contributing to your decision trees. This could (and often does) imply that those features are the most important predictors of kidney disease. #### Question 7

The Random Forest is an "ensemble" method (like bagging) that bootstraps many decision trees using the same training data and uses the average predictions of these trees to classify test observations. By averaging over many potentially different trees we derive a bigger picture of the associations in the data than a single tree could give us. Importantly, what differentiates Random Forests from Bagging is that they only consider a random subset of features in each bootstrapped tree (the `sklearn` default subset size is the square root of the the number of features). There are two reasons to do this: 

1) If we have correlated features in a tree and at least some of them are important, then bootstrapped trees will tend to be very similar to each other because the set or sets of correlated features will dominate the tree's decisions. By taking a random subset of features that may or may not include correlated features, the trees themselves become less correlated and therefore less biased toward certain decision boundaries.

2) By decorrelating trees, we can derive independent estimates of each variable's average contribution to node purity (reduction in G).

Fit a random forest classifier to your training data.

```python
from sklearn.ensemble import RandomForestClassifier

forest = RandomForestClassifier(n_estimators=1000, random_state=42)
forest.fit(X_train, y_train)
```

Create a sorted barplot of `forest.feature_importances_`, with feature names on one axis. The feature importances are the average total decrease in node impurity for a feature averaged across all trees fit in the random forest.

```python
importances = forest.feature_importances_
importance = pd.DataFrame({"feature": features.columns, "importance": importances})
importance_sorted = importance.sort_values("importance")

# Create a sorted barplot
sns.barplot(x="importance", y="feature", data=importance_sorted)
```
